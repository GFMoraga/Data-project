```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import xarray as xr

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

%matplotlib inline
plt.rcParams.update({'font.size': 12})
```

```python
np.random.seed(0)

mu = 3
n=50

x = np.random.normal(mu, 1, size=n)
y = np.random.normal(mu, 1, size=n)
z = np.random.normal(mu, 1, size=n)
```

```python
plt.rcParams['figure.figsize'] = (8,6)
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, s=50);

# Labeling

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Tilting
# Degrees

ax.view_init(45, 215)
```

```python
# 3D LINE plot

omega = 2

z_line = np.linspace(0, 10, 100)
x_line = np.cos(omega*z_line)
y_line = np.sin(omega*z_line)
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(x_line, y_line, z_line)
```

```python
# Contour Plot

# Set F

f = lambda x, y: np.sin(x) * np.cos(y)
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(x, y)
F = f(X, Y)
```

```python
plt. contour(X, Y, F)
plt.show()
```

```python
f = lambda x, y: np.sin(x) * np.cos(y)
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
F = f(X, Y)

fig = plt.figure(figsize = [12, 8])
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, F, cmap=cm.coolwarm)

plt.show()
```
